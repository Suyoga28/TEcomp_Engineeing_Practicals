```python
def remove_outliers_iqr(dataf, column):
    Q1 = dataf[column].quantile(0.25)    #25% of the values are below this
    Q3 = dataf[column].quantile(0.75)    #75% of the values are below this
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR    # below lower_bound or above upper_bound is considered an outlier
    upper_bound = Q3 + 1.5 * IQR
    filtered_data=dataf[(dataf[column] >= lower_bound) & (dataf[column] <= upper_bound)]
    return filtered_data

numeric_cols = df.select_dtypes(include=np.number).columns  #For categorical data, the concept of "outliers" doesn't apply cuz of continuous scale
for col in numeric_cols:
    df = remove_outliers_iqr(df, col)

print("Shape after outlier removal:", df.shape)
```

```python
col = ['math score', 'reading score', 'writing score', 'placement score']
df.boxplot(column=col)
plt.show()
```

```python
from sklearn import preprocessing
scaler = preprocessing.MinMaxScaler()
df_scaled = pd.DataFrame(scaler.fit_transform(df[numeric_cols]), columns=numeric_cols)  

#ensures that the column names in df_scaled are the same as the original numeric columns in df

print("Scaled Data (first 5 rows):")
print(df_scaled.head())
```

# The log transformation is used in data preprocessing to make skewed data more normally distributed.

# Histogram of original scores
#df['math score'].plot(kind='hist', title='Original Math Score')
#plt.show()


```python
import seaborn as sns
sns.histplot(data=df['math score'], bins=10, kde=False, color='blue')
plt.title("Original Math Score")
```

```python
# Log transformation (handling zero values safely)  Log of 0 or negative numbers is undefined.
df['log_math'] = np.log10(df['math score'] + 1)    #Make data more "normal" (bell-shaped) from skewed
```


# Histogram of log-transformed scores
df['log_math'].plot(kind='hist', title='Log-Transformed Math Score')
plt.show()

```python
sns.histplot(data=df['log_math'], bins=10, kde=False, color='blue')
plt.title("Log_transformed Math Score")
```

```python
plt.figure(figsize=(8, 4))
sns.boxplot(data=[df['math score'], df['log_math']])
plt.xticks([0, 1], ['Original', 'Log Transformed'])
plt.title("Boxplot Comparison")
```

```python
# OPTIONAL

pt = preprocessing.PowerTransformer(method='yeo-johnson')
df_normalized = pd.DataFrame(pt.fit_transform(df[numeric_cols]), columns=numeric_cols)

df_normalized.hist(bins=30, figsize=(12, 8))
plt.suptitle("Normalized Feature Distributions")  # places a title at the top center of the full figure — not just one subplot.
plt.tight_layout()
plt.show()
```
