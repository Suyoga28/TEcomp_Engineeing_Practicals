```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import sklearn
from sklearn.datasets import fetch_california_housing
```

```python
housing=fetch_california_housing()
data=pd.DataFrame(housing.data)
```

```python
data.columns=housing.feature_names
```

```python
data["PRICE"]=housing.target
```

```python
x=data.drop("PRICE",axis=1)
y=data["PRICE"]
```

```python
from sklearn.model_selection import train_test_split
```

```python
xtrain,xtest,ytrain,ytest= train_test_split(x,y,test_size=0.2,random_state=0)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
lm=LinearRegression()

model=lm.fit(xtrain,ytrain)
```

```python
ytrain_pred=lm.predict(xtrain)
ytest_pred=lm.predict(xtest)
```

```python
df_train=pd.DataFrame({"Actual":ytrain,"Predicted":ytrain_pred})
df_test=pd.DataFrame({"Actual":ytest,"Predicted":ytest_pred})
```

```python
from sklearn.metrics import mean_squared_error,r2_score
```

```python
mse=mean_squared_error(ytrain,ytrain_pred)
print(mse)
mse=mean_squared_error(ytest,ytest_pred)
print(mse)
```

```python
plt.scatter(ytrain,ytrain_pred,color="green",marker='d',label="Training data")
plt.scatter(ytest,ytest_pred,color="blue",marker='*',label="Testing data")
plt.title("Actual value VS Predicted values")
plt.xlabel("Actual Values")
plt.ylabel("Predicted Values")
plt.legend(title="Actual vs Predicted")

min_value=min(ytrain.min(),ytest.min())
max_value=max(ytrain.min(),ytest.max())

plt.plot([min_value,max_value],[min_value,max_value],'k--',lw=2)
```
