```python
#Install	pip install scikit-learn
#Import	   from sklearn import preprocessing

from sklearn import preprocessing
min_max_scaler = preprocessing.MinMaxScaler()   # creating a object of the MinMaxScaler class to implement method - fit, transform
```

```python
x = df.iloc[:, :4]   #integer-location based indexing
#           |    |
#          rows  cols      so selecting first 4 cols
```

```python
x
```

```python
x_scaled = min_max_scaler.fit_transform(x)

#min_max_scaler: This is an obj of MinMaxScaler from sklearn.preprocessing. It’s used to scale/normalize the data to a range — typically [0, 1].

#fit(x): Calculates the minimum and maximum values for each column in x.
#transform(x): Scales the data based on those minimum and maximum values.

#    X scaled = X​ −X min 
#             -------------             
#              X max −X min


#Appling it separately
#scaler.fit(df)
#df_scaled = scaler.transform(df)   always save in var for transform

```

```python
df_normalized = pd.DataFrame(x_scaled)
# Takes the scaled data from x_scaled (which is a numpy array) and converts it into a DataFrame

df_normalized
```

Using label Encoder Now Converting categorical variable into numerical -- required order, so applying for only species column

```python
label_encoder = preprocessing.LabelEncoder() 
#encode categorical labels into numerical values and works if there is an actual order like high, medium, low otherwise one-hot encoder

df['species'] = label_encoder.fit_transform(df['species'])

#fit(): This method learns the unique categories (labels) in the column and assigns them an integer.
#transform(): This method replaces the categorical values with corresponding integers.
```

```python
df['species'].unique()
```

```python
features_df = df.drop(columns=['species']) 
# new DataFrame (features_df) withoout species col
#you may want to separate features (the independent variables) from the target variable (the dependent variable).
```




Using One hot Encoder Now Converting categorical variable into numerical -- required order, so applying for only species column




```python
enc = preprocessing.OneHotEncoder()  
#useful when you have categorical features with no order and encode categorical labels into binary not numbers 
#Result: This will give you a sparse matrix of encoded values

enc_df=pd.DataFrame(enc.fit_transform(df[['species']]).toarray())
#toarray(): Converts the sparse matrix returned by fit_transform() into a regular NumPy array
```

#Before feeding them into a machine learning model

```python
df_encode = features_df.join(enc_df)
# joins the features_df DataFrame with enc_df along the index  combining different types of features into a single DataFrame


df_encode.head()
```

```python
df_encode.rename(columns = {0:'Iris-Setosa',
                            1:'Iris-Versicolor',
                            2:'Iris-virginica'}, inplace = True)


# Inplace- It tells pandas to make the changes directly to the DataFrame (i.e., modify it "in place") without returning a new copy
```

```python
one_hot_df = pd.get_dummies(
    df,                      # original DataFrame
    prefix="Species",        # prefix for new column names
    columns=['species'],     # column(s) to encode
    drop_first=True          # drop the first category to avoid 'multicollinearity'- contain similar information about the variance in the target
                             # (dependent) variable. It makes it hard to determine the effect of each independent variable.
)

#convert categorical variables into dummy/indicator variables (also called one-hot encoding).
# Pandas function returs whole dataframe not sparse matrix like OneHotEncoder
```
