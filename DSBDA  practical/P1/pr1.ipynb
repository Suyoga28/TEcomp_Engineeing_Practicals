```python
dff=df.iloc[:,:4]
```

```python
min_max=preprocessing.MinMaxScaler()
df_scaled=min_max.fit_transform(dff)
```

```python
df_normalised=pd.DataFrame(df_scaled)
```

```python
df_normalised.head()
```

```python
df_normalised=np.ceil(df_normalised).astype(int)
```

```python
df_normalised.head()
```

```python
label_encoder=preprocessing.LabelEncoder()
df["species"]=label_encoder.fit_transform(df["species"])
#df["species"]=pd.DataFrame(df["species"])
df["species"].head()
```

```python
first_df=df.drop(columns=["species"])
one_hot_enc=preprocessing.OneHotEncoder()
df_enc=pd.DataFrame(one_hot_enc.fit_transform(df[["species"]]).toarray())

```

```python
df_fin=first_df.join(df_enc)
df_fin.head()
```

```python
df_fin.rename(columns={
                        0:"Iris-Setosa",
                        1:"Iris-Ve"
})
```

```python
df_df=pd.get_dummies(df,prefix="Species",columns=["species"],drop_first=True)
```
