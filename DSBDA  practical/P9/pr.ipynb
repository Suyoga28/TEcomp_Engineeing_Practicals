```python
import pandas as pd 
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv("train.csv")
df.head()
```

```python
df.size
```

```python
sns.boxplot(x="Sex",y="Age",data=df)
```

```python
sns.boxplot(x="Survived",y="Fare",data=df)
```

```python
print(df["Survived"].unique())
```

```python
df1=df.dropna(subset=["Age","Sex"])
sns.boxplot(x="Sex",y="Age",hue="Survived",data=df1)
plt.title("Distribution of age according to geneder and survival rate")
plt.xlabel("Sex")
plt.ylabel("Age")
plt.legend(title="Survived",labels=["No","Yes"])
plt.show()
```

```python
plt.figure(figsize=(12,6))
sns.histplot(
    x="Age",
    hue="Sex",
    data=df,
    stat="density",
    common_norm=False,
    kde=True,
    alpha=0.5
)
plt.title("HISTOGRAM")
plt.xlabel("Sex")
plt.ylabel("Frequency");
```
