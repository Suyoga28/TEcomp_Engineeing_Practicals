```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

df=pd.read_csv('train.csv')
df
```

```python
sns.pairplot(df)
```

```python
sns.distplot(df["Fare"],kde=False,color="darkgreen")
```

```python
sns.rugplot(df["Fare"],height=0.1,palette="muted")
plt.ylabel("Frequency")
```

```python
sns.jointplot(x="Age",y="Fare",data=df,kind="scatter",color="pink")
```

```python
sns.barplot(x="Sex",y="Fare",data=df,estimator=np.std,palette="Purples")
```

```python
sns.boxplot(x="Sex",y="Fare",data=df,hue="Survived")
plt.legend(title="Boxplot",loc="center")
```

```python
plt.figure(figsize=(16,7))
sns.countplot(x="Fare",data=df,palette="rainbow")
plt.xticks(rotation=45);
```

```python
sns.violinplot(x="Sex",y="Fare",data=df,palette={"male":"skyblue","female":"lightpink"})
```

```python
sns.stripplot(x="Sex",y="Fare",data=df,hue="Survived",jitter=0.1)
```

```python
sns.histplot(x="Pclass",y="Survived",data=df,hue="Survived",palette="Oranges")
```
