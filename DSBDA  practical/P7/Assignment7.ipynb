```python
# Natural Language Toolkit

import nltk  
from nltk.tokenize import sent_tokenize  #Sentence Tokenisation -split the text into sentences
#nltk.download('punkt')  already downloded

text="""Hello Mr. Smith, how are you doing today ? The weather is great, and the city is awesome ! The sky is pinkish-blue. You shouldn't eat cardboard """

tokenized_text=sent_tokenize(text)
print(tokenized_text)
```

```python
from nltk.tokenize import word_tokenize     #Word Tokenisation

tokenized_word=word_tokenize(text)
print(tokenized_word)
```

```python
from nltk.probability import FreqDist
#counting how often each word/token appears
#tokenized_word should be a list of words (not sentences)

fdist = FreqDist(tokenized_word)
print(fdist)
print("\nItems")
print(fdist.items()) # All word-frequency pairs

print("\nCommon words")
print(fdist.most_common(5)) # Top 5 most common words




# 27 unique tokens -> samples      31 total tokens (including duplicates)-> outcomes
```

```python
from nltk.corpus import stopwords #don't help in understanding the main content 

#nltk.download('stopwords')  already downloaded
stop_words=set(stopwords.words("english"))
print(stop_words)
```

```python
filtered_sent=[]
for w in tokenized_word:
    if w not in stop_words:
        filtered_sent.append(w)
        
print("Tokenized Sentence:",tokenized_word)
print("\nFilterd Sentence:",filtered_sent)
```

```python

from nltk.stem import PorterStemmer
from nltk.tokenize import sent_tokenize, word_tokenize

ps = PorterStemmer()

stemmed_words=[]
for w in filtered_sent:
    stemmed_words.append(ps.stem(w))

print("Filtered Sentence:",filtered_sent)
print("Stemmed Sentence:",stemmed_words)
```

```python
from nltk.stem.wordnet import WordNetLemmatizer

#nltk.download('wordnet') already downloaded
lem = WordNetLemmatizer()

word = "flying"
print("Lemmatized Word:",lem.lemmatize(word,"v"))  #Reduces the word to its base form (lemma) using correct part of speech (POS).
                                                   #"v" means verb, so "flying" → "fly".

print("Stemmed Word:",ps.stem(word)) #simply cuts off suffixes using crude rules
```

```python
import nltk
import pandas as pd
data=pd.read_csv('train.tsv' , sep='\t')  # just need to tell it what the delimiter
data.head()
```

```python
from sklearn.feature_extraction.text import CountVectorizer
from nltk.tokenize import RegexpTokenizer

#CountVectorizer: Turns text into a word-count matrix.
#RegexpTokenizer: A custom tokenizer that splits text using regular expressions.

token = RegexpTokenizer(r'[a-zA-Z0-9]+')  #keeps only words and numbers, removing punctuation.

cv = CountVectorizer(
    lowercase=True,                 # Converts all words to lowercase
    stop_words='english',           # Removes common stopwords (like "the", "and", "is")
    ngram_range=(1, 1),             # Uses single words only (unigrams)
    tokenizer=token.tokenize        # Uses your custom tokenizer
)

text_counts= cv.fit_transform(data['Phrase'])

#Result: A sparse matrix where:

#Each row = one phrase
#Each column = one word
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(
    text_counts, data['Sentiment'], test_size=0.3, random_state=1)
```

```python
from sklearn.naive_bayes import MultinomialNB
from sklearn import metrics

clf = MultinomialNB().fit(X_train, y_train)
xtest_pred= clf.predict(X_test)
print("MultinomialNB Accuracy:",metrics.accuracy_score(y_test, xtest_pred))

#multinomial distribution-it models the number of times each outcome happens 
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer   #turning text into numbers 

tf=TfidfVectorizer()
text_tf= tf.fit_transform(data['Phrase'])

#TF = Term Frequency → How often a word appears in a document.
#IDF = Inverse Document Frequency → How rare the word is across all documents.
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(
    text_tf, data['Sentiment'], test_size=0.3, random_state=123)
```

```python
#Model generation

clf = MultinomialNB().fit(X_train, y_train)
ytest_pred= clf.predict(X_test)
print("MultinomialNB Accuracy:",metrics.accuracy_score(y_test, ytest_pred))
```
