Compute and display summary statistics for each feature available in the dataset.

```python
import seaborn as sns
import matplotlib
import matplotlib.pyplot as plt

%matplotlib inline
#It tells the notebook to display plots directly below the code cell that produced them.
#Without this, your plots may not show up (or may pop up in a new window).

```

Data Visualization-Create a histogram for each feature in the dataset to illustrate the feature distributions. Plot each histogram.

```python
fig, axes = plt.subplots(2, 2, figsize=(16, 8))   #plt.subplots() creates a grid of plots 2, 2 means 2 rows and 2 columns (4 plots), figsize optional


''' subplot return 2 things:
fig = a big piece of paper (the whole figure)
axes = 4 little drawing areas -on that paper
Each axes[i][j] = a specific spot where you can draw your graph '''


axes[0,0].set_title("Distribution of First Column")
axes[0, 0].hist(df["col1"], bins=20, color='skyblue', edgecolor='black')   #bins means no. of columns adds detailing 
axes[0, 0].set_xlabel('Value')
axes[0, 0].set_ylabel('Frequency')

axes[0,1].set_title("Distribution of Second Column")
axes[0,1].hist(df["col2"])

axes[1,0].set_title("Distribution of Third Column")
axes[1,0].hist(df["col3"])

axes[1,1].set_title("Distribution of Fourth Column")
axes[1,1].hist(df["col4"])

plt.tight_layout()  # At last to avoid overlapping

plt.show() # OPTIONAL IN NB
```

```python
#sns.set_style("whitegrid")  #sets background

# Creating a figure instance
fig = plt.figure(1, figsize=(12,8))  # 1 is fig number

ax = fig.add_subplot(111)
# Boxplot can be with help of subplot(need store var) or sns.
# 111 is a shorthand for a grid layout: 1 row, 1 column, and this is the 1st subplot.

# Creating the boxplot
data_to_plot = [df["col1"],df["col2"],df["col3"],df["col4"]]
ax.boxplot(data_to_plot)

ax.set_title('Boxplot of Selected Columns')
ax.set_ylabel('Values')
ax.set_xticklabels(['Column 1', 'Column 2', 'Column 3', 'Column 4'])
#Tick labels are the names or values next to little marks on axis.
```

Create a boxplot for each feature in the dataset. All of the boxplots should be combined into a single plot. Compare distributions and identify outliers.



If we observe closely. for the box 2, interquartile distance is roughly around 0.75 hence the values lying beyond this range of (third quartile + interquartile distance) i.e. roughly around 4.05 will be considered as outliers. Similarly outliers with other boxplots can be found

```python
#Seaborn (built on top of Matplotlib) so automatically group and label
plt.figure(figsize=(10, 4)) 
sns.boxplot(x='col1', y='col2', data=df) #sns.boxplot(...) auto-sets the labels from column names.

#boxplot = sns.boxplot(x='col1', y='col2', data=df)  for customise label you have to store sns.boxplot in boxplot var
#boxplot.set_xlabel("My Custom X Label")
#boxplot.set_ylabel("My Custom Y Label")

plt.xticks(rotation=45);  # NOTE Semicolon to avoid unwanted output
```
