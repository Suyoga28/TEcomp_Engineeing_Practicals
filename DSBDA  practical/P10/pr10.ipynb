```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("iris-flower-dataset (1).csv")
```

```python
df.columns=["col1","col2","col3","col4","col5"]
```

```python
df.info()
```

```python
df.describe()
```

```python
fig,axes=plt.subplots(2,2,figsize=(14,6))
axes[0][0].hist(df["col1"],color="skyblue",bins=20,edgecolor="black");
axes[0][0].set_title("MY Graph")
axes[0][0].set_xlabel("Value")
axes[0][0].set_ylabel("Frequency")

axes[0][1].hist(df["col2"],color="skyblue",bins=20,edgecolor="black");
axes[0][1].set_title("MY Graph")
axes[0][1].set_xlabel("Value")
axes[0][1].set_ylabel("Frequency")

plt.tight_layout()
```

```python
sns.set_style("whitegrid")
```

```python
fig=plt.figure(1,figsize=(12,6))
ax=fig.add_subplot(111)
box_data=[df["col1"],df["col2"],df["col3"],df["col4"]]
ax.boxplot(box_data);
ax.set_title("Boxplot")
ax.set_xticklabels(["col1","col2","col3","col4"])
```

```python
plt.figure(figsize=(12,6))
sns.boxplot(x="col1",y="col2",data=df)
plt.xticks(rotation=45);
```
